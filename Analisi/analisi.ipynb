<h1 style="color: blue; font-size: 50px;"> Confronto fra configurazioni </h1>

```python
fig, ax = plt.subplots(1, 2, figsize = (15, 7))

conf1 = np.loadtxt("condin.conf")
conf2 = np.loadtxt("conf100.conf")

ax[0].scatter(conf1[:, 1] * np.cos(conf1[:, 2]), conf1[:, 1] * np.sin(conf1[:, 2]), color = "darkblue")
ax[0].set_title("Configurazione iniziale")

ax[1].scatter(conf2[:, 0], conf2[:, 1], color = "darkred")
ax[1].set_title("Configurazione evoluta")

plt.show()
```
